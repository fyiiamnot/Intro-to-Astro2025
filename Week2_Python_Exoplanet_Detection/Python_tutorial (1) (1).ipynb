```python
#loading the image
#Copy path and change it to load
from astropy.io import fits
file_dir = "C:/Users/Tshewang Phuntsho/Desktop/test_data/"
img_name = 'pdi_pi_collapsed.fits'
img = fits.getdata(file_dir+img_name,ext=0)
```

```python
# Display the img
img_corrected = np.transpose(img, (1, 2, 0))  # change [c,h,w] to [h,w,c]
plt.imshow(img_corrected, origin='lower', vmin=-0.001, vmax=0.01)
plt.colorbar()
plt.show()
```

```python
# create a figure
fig = plt.figure(figsize=(8,8))
img = np.transpose(img, (1,2,0))
# Show the image 
image = plt.imshow(img, origin = 'lower', vmin=-0.001, vmax=0.01)
cbar = fig.colorbar(image,shrink=0.82)
cbar.set_label('Surface Brightness (arb. units)', rotation=90,fontsize=12)
plt.xlim([60,140])
plt.ylim([60,140])
plt.xlabel(r'$\Delta$ RA [px]',fontsize=12)
plt.ylabel(r'$\Delta$ Dec [px]',fontsize=12);
```

```python
# Reading 20,000 Rows*97 columns of Data- Closest 20k stars from Gaia Archive (DR2)
# Downloaded the csv file from the Gagia database 
file_dir = "C:/Users/Tshewang Phuntsho/Desktop/csv_files_for_what/"
stellar=pd.read_csv(file_dir + "closest20kstars.csv")

# Creating a matplotlib (pyplot) figure
fig = plt.figure(figsize = [10,10])

# Plotting the scatter graph with RA on x-axis and Dec on y-axis. alpha tells the opacity, s is the size
plt.scatter(stellar['ra'], stellar['dec'], alpha=0.8, s=12, lw = 0.5, ec = 'k', color='darkorchid')
# Define the characteristics of the plot
plt.xlim([0,362])
plt.ylim([-75,75])
plt.xlabel(r'RA [$\degree$]', fontsize = 14)
plt.ylabel(r'Dec [$\degree$]', fontsize = 14)
plt.title('Scatterplot of Closest Stars 20,000 stars from Gaia Archive (DR2)', fontsize = 16)
plt.show()
plt.savefig("my_scatterplot.png")
```
